## __plot 그리기__ <a name="p1section1"></a>
필요 라이브러리를 로드 합니다.  
데이터 관리를 위한 numpy 라이브러리가 함께 필요합니다.  
(note : notebook 파일은 셀 내 그림이 표현되는 반면, python 파일은 그림을 위한 window가 별도로 생성되는 면에서 일부 다릅니다.)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(0, 5, 10) # 0부터 5까지 10개의 점 생성
ys = xs**2 # numpy ndarray(N dimansional array) 전체에 대한 제곱 연산 
print(xs)
print(ys)
```

```python
plt.figure() # 그림 작성을 위한 기본판 생성
plt.plot(xs, ys, 'r') # 'r' : red color line 으로 그리기
```

그림에 축 설정등은 아래와 같이 더할 수 있습니다.

```python
plt.xlabel('x') # 축 라벨 성정
plt.ylabel('y')
plt.title('title') # 그림 타이틀 설정
plt.plot(xs, ys, 'r')
plt.show() # python 파일에서는 그림판 보이기를 해야 실제 그림이 나타납니다.
```

여러 개의 그림을 표현하려면 subplot 의 개념을 씁니다.  
그림 1개는 Axes 라는 클래스(설계도)를 통해 정의되고 각 subplot 이 Axes 클래스의 instance에 해당합니다.  
(X축, Y축으로 이루어져서 Axes 라는 이름이 붙었습니다.)

```python
fig1 = plt.figure() # python 파일 실행 시 별도의 그림을 위한 window 를 생성합니다.
axes1 = fig1.add_subplot(2,2,1) # 가로 2, 세로 2 개의 타일 형태의 subplot 을 만들고 1번째 위치의 subplot 으로 지정
axes2 = fig1.add_subplot(2,2,4) # 4번째 위치 (행, 렬에 대해 차례대로 번호를 붙였을때)
axes1.plot(xs, ys, 'g') # 'green'
axes1.plot(xs, ys+2, 'r') # 'red'
axes2.plot(xs, ys*2, 'y') # 'yellow' line at axes2
plt.show()
```

## __다양한 종류의 plot__ <a name="p1section2"></a>

```python
# 데이터 추출을 위한 랜덤함수를 호출합니다
from numpy.random import randn
xs2 = randn(50) # normal 분포로 부터 50개의 랜덤 실수를 numpy ndarray 형태로 생성
ys2 = randn(50)

fig2 = plt.figure()
axes21 = fig2.add_subplot(2,2,1)
axes22 = fig2.add_subplot(2,2,2)
axes23 = fig2.add_subplot(2,2,3)
axes24 = fig2.add_subplot(2,2,4)

# 몇 가지 형태로 그림 그리기
axes21.hist(xs2)
axes22.scatter(xs2, ys2)
axes23.plot(xs2.cumsum(), ys2.cumsum(), '--') # cumsum : 데이터에 대하여 이전 값부터 누적 sum 데이터 값 도출, '--' : 긴 점선
axes24.bar(xs2, abs(ys2), width=0.1) # abs() : 절대값, width : bar 그림 두께
plt.show()
```

```python
# 그림에 대한 기타 옵션
fig3 = plt.figure()
plt.plot(xs, ys) # 여러 그림을 그리면서 color 지정 하지 않을 시 자동 변경
plt.plot(xs, ys+2, marker='+') # 점 '+' 형태로 표시, ys+2 : ys 각 원소에 2 더하기
plt.plot(xs, ys+4, marker='+', markersize=4)
plt.plot(xs, ys+6, marker='+', markersize=4, linestyle='--')
plt.plot(xs, ys+8, marker='+', markersize=4, linestyle='steps-post') # 계단형 선
plt.plot(xs, ys+10, marker='+', markersize=4, linestyle='--', linewidth=4) # line 두께 변경
plt.plot(xs, ys+12, marker='s', markersize=8, linestyle='--', linewidth=4, markerfacecolor='yellow', color='k', 
         markeredgecolor='red', markeredgewidth=3) # marker color 변경
plt.plot(xs, ys+14, marker='*', linestyle='none') # 선없이 점만 그리기
plt.show()
```

## __Plot 기타 작업__ <a name="p1section3"></a>

홈페이지에 있는 figure 의 기본 성질입니다.  
세부적인 사용 방법 등이 적혀 있으며 몇 가지 예시를 보면 아래와 같습니다.  
<img src="figure.PNG" width="600"/>

위 class 설명을 보면  
class matplotlib.figure.Figure(figsize=None, dpi, ...) : 클래스가 figsize, dpi, ... 같은 parameters를 받을 수 있고,  
각 parameters 설명을 보면  
figsize는 그림판의 크기 변경을 적용합니다.  
두 개의 실수 값으로 주어진 tuple 로 (width, height) 형식으로 주어집니다.  
dpi 는 그림의 해상도를 결정합니다.  

```python
fig1 = plt.figure(figsize=(16, 9)) # 16:9 크기로 설정
ax1 = fig1.add_subplot(1,1,1)
ax1.set_title('Title is here')
ax1.set_ylim([0, 10]) # x축 범위 설정
ax1.set_xlim([-1, 5])
ax1.set_xlabel('x axis') 
ax1.set_ylabel('y axis')
ax1.plot(xs, ys, label='plot 1') # 레전드를 생성하기 위한 라벨 설정
ax1.plot(xs, ys*0.5, label='plot 2')
ax1.set_xticks([-1, 1, 3, 5]) # X축에 값을 표시할 눈금 위치 설정
ax1.set_xticklabels(['label1', 'label2', 'label3', 'label4']) # X축에 눈금 값 변경
ax1.legend(loc='best') # legend 를 위치 자동으로 추가
plt.show()
```

### __축 조절__ <a name="p1section31"></a>

```python
# figure 생성과 axes 생성을 동시에 할 수 있습니다.
# 위에서와 같이 figure 생성 후 axes 별도로 추가하지 않고 한 번에도 처리 가능합니다
# 가로 1, 세로 2 개의 그림을 만들고 각각의 그림은 ax[0], ax[1] 에 저장합니다.
fig3, ax = plt.subplots(1,2, figsize=(16, 9))
ax[0].plot(xs**3, ys)
ax[0].set_yscale('log') # y축 값을 로그스케일로 표시합니다.
ax[0].grid(True) # 그림에 격자를 표시합니다.
ax[1].plot(xs**3, ys)
ax[1].set_yscale('log')
ax[1].grid(color='0.3', alpha=0.8, linestyle='dashed', linewidth=0.3) # alpha 투명도
plt.show()
```

### __색깔__ <a name="p1section32"></a>
선 뿐만 각 점들도 개별적인 색깔 적용이 가능합니다.  
직접적인 RGB값을 넣거나 사전 정의된 값들을 사용 가능합니다.

```python
fig1, ax1 = plt.subplots(1,1)
xs = range(1, 11) # 1, 2, 3, ..., 10
ys = np.array(xs)*2 # Y = 2 * X
# 10개의 점들에 대한 RGB color 값을 랜덤하게 생성
colors = np.random.rand(10, 3) # [[r1, g1, b1], [r2, g2, b2], ..., [r10, g10, b10]], rgb in [0, 1]
ax1.scatter(xs, ys, c=colors) # c=colors : 색깔 값에 colors 변수 값을 적용
plt.show()
```
