### __Plot 옵션__ <a name="r3section11"></a>
plot 주요 parameter  
* type : 그림 타입(선, 점, ...)
* xlim, ylim : 그림 범위
* log : 로그 스케일 'x', 'y', 'xy'
* main : title
* sub : subtitle
* xlab, ylab : 축에 대한 label

```r
plot(hills$dist, xlab='x axis', ylab='y axis') # hills의 dist 값들을 x축(index) 에 대해 plot
plot(hills$climb ~ hills$dist ) # 두 변수를 축으로 plot
```

```r
plot(hills$climb ~ hills$dist, type='p' ) # 점으로 표시
plot(hills$climb ~ hills$dist, type='l' ) # 선으로 표시
plot(hills$climb ~ hills$dist, type='b' ) # 점 + 선
plot(hills$climb ~ hills$dist, type='s' ) # 계단식
plot(hills$climb ~ hills$dist, type='h' ) # horizontal
plot(hills$climb ~ hills$dist, type='n' ) # 점, 선 없음 (여러 그림들을 겹쳐 그릴 때 같은 축 크기로 맞출 수 있음)
```

```r
plot(hills$climb ~ hills$dist, xlim = c(-10, 30), ylim = c(0, 10000), main='위에 제목', sub='밑에 subtitle' )
```

### __Bar chart__ <a name="r3section13"></a>

```r
barplot(hills$dist) # 위 plot 의 점을 bar로 표시
```

```r
options(repr.plot.width=6, repr.plot.height=5) # 그림 크기 설정
barplot(hills$dist, horiz=TRUE, # horizontal 방향으로 그리기
       space=1, # bar 간 빈공간
        names.arg = (1:35), # bin에 이름 적용
        col = 'blue'
       )
```
